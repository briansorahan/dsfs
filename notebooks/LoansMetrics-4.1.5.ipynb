```python
import pandas                   as pd
import numpy                    as np
import sklearn.cross_validation as cv
import sklearn.metrics          as metrics

from sklearn import linear_model

df = pd.read_csv('loansData.csv')

df.dropna(inplace=True)

# Create the interest rate and annual income series.
df['int_rate'] = df['Interest.Rate'].map(lambda val: float(val[:-1]))
df['annual_inc'] = df['Monthly.Income'] * 12

df.head()
```

```python
# Predict interest rate from annual income.

est = linear_model.LinearRegression()
X   = df['annual_inc'].reshape(-1, 1)
y   = df['int_rate']
mae = cv.cross_val_score(est, X, y, scoring='mean_absolute_error', cv=10)
mse = cv.cross_val_score(est, X, y, scoring='mean_squared_error', cv=10)
r2  = cv.cross_val_score(est, X, y, scoring='r2', cv=10)

print('average mae is {}'.format(np.mean(mae)))
print('average mse is {}'.format(np.mean(mse)))
print('average r2  is {}'.format(np.mean(r2)))
```
